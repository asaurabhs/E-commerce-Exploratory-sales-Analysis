# visual data analysis

```python
#To plot a bar graph based on Gender market share
sns.countplot(x = 'Gender', data = data)
```

```python
ax = sns.countplot(x = 'Gender', data = data)
for i in ax.containers:
    ax.bar_label(i)
```

```python
# to plot a bar based on Gender Expenditure
data.groupby(['Gender'], as_index = False)['Amount'].sum()
```

```python
market_share = data.groupby(['Gender'], as_index = False)['Amount'].sum()
```

```python
sns.barplot(x = 'Gender', y = 'Amount' ,data = market_share)

```

```python
# I want to simply filter out customers based on their Gender and age group
sns.countplot(x = 'Age Group', hue = "Gender", data = data)
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = data)
for i in ax.containers:
    ax.bar_label(i)
```

```python
#bar plot showing total amount sold for each product category

category = data.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
ax = sns.barplot( x = 'Product_Category', y = 'Amount', data = category, palette = "dark:salmon_r")
sns.set(rc = {'figure.figsize':(20,20)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
data.groupby(['Occupation'], as_index = False)['Amount'].sum()
```

```python
#bar plot showing total amount spent by males and females
gen_shop = data.groupby(['Gender'], as_index = False)["Amount"].sum()
```

```python
ax = sns.barplot(x = 'Gender', y ='Amount', data = gen_shop, palette = "husl")
for i in ax.containers:
    ax.bar_label(i)


```

```python
#Barplot regarding customers occupation (what do they do for a living)
occupation_details = data.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)
```

```python
sns.barplot(x = 'Occupation', y = 'Amount', data = occupation_details,palette = "icefire")
sns.set(rc = {'figure.figsize':(20,10)})
```

```python
data.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = "Amount", ascending = False)
```

```python
top_products = data.groupby(['Product_Category','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
```

```python
#stacked bar plot showing Total amount sold for each product category and  gender
table = top_products.pivot(index='Product_Category', columns='Gender', values='Amount')

#stacked bar chart of the total amount sold for each product and gender
table.plot(kind='bar', stacked=True)

# Set the title of the plot
plt.title('Top 10 Products by Sales and Gender')

# Set the x-axis label
plt.xlabel('Product')

# Set the y-axis label
plt.ylabel('Total Amount Sold')

# Show the plot
plt.show()
```

```python
#stacked bar plot showing the total amount sold for each product category and marital status.

demo = data.groupby(['Product_Category', 'in_relationship'],as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)

table = demo.pivot(index = 'Product_Category', columns = 'in_relationship', values = 'Amount')
table.plot(kind = 'bar', stacked = True)
```

```python
#Create a grouped bar plot showing the total amount sold for each product category and gender.
demo1 = data.groupby(['Product_Category', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
table = demo1.pivot(index= 'Product_Category', columns = 'Gender', values = 'Amount')
table.plot(kind = 'bar', stacked = True)
```
